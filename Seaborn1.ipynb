```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
titanic
titanic.head()
```

```python
sns.jointplot(data=titanic, x='fare', y='age')
sns.set_style('whitegrid')
```

```python
sns.distplot(titanic['fare'], kde=False, bins=30, color='r')
```

```python
titanic.head()
```

```python
sns.boxplot(data=titanic, x='class', y='age', palette='rainbow')
sns.set_style('whitegrid')
```

```python
sns.swarmplot(data=titanic, x='class', y='age')
sns.set_style('whitegrid')
```

```python
sns.countplot(data=titanic, x='sex')
```

```python
tc = titanic[['survived','pclass','age','sibsp','parch','fare','adult_male','alone']]
tcc = tc.corr()
sns.heatmap(tcc, cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
g = sns.FacetGrid(data=titanic,col='sex')
g.map(sns.histplot,'age')
```
